```python
import pandas as pd
import numpy as np
```

```python
df_can = pd.read_excel('Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=(20),
                       skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.tail()
```

```python
df_can.info()
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace=True)
```

```python
df_can.columns
```

```python
df_can['Total']=df_can.sum(axis=1)
df_can.head()
```

```python
df_can.isnull().sum()
```

```python
df_can.dropna(axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.__version__
```

```python
print(plt.style.available)
```

```python
years = range(1980,2013)
years
```

```python
df_ind_sg = df_can.loc[['Indonesia', 'Singapore'], years]
```

```python
df_can_sorted = df_can.sort_values(by='Total', axis=0, ascending=False)
df_top5 = df_can_sorted.head(5)
df_top5
```

```python
df_top5.plot(kind='line', figsize=(28,16))
df_top5.index = df_top5.index.map(int)

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Top 5 Countries Immigrating to Canada 1980-2013')
```

```python
df_least5 = df_can.sort_values(by="Total", axis=0).head(5)
df_least5 = df_least5[years]

df_least5
```

```python
df_can_sorted = df_can.sort_values(by='Total', axis=0, ascending=False)
df_can_sorted[2013]
```

```python
df_dns = df_can_sorted.loc[['Denmark','Norway','Sweden'], years].transpose()

count, bin_edges = np.histogram(df_dns, bins=20)

df_dns.plot(kind="hist", bins=20, xticks=bin_edges, figsize=(16,8), alpha=0.3, color=["red","blue","green"])

plt.title("Denmark, Norway, Sweden immigration to Canada 1980-2013")
plt.ylabel("Number of Years")
plt.xlabel("Number of Immigrants")

plt.show()
```

```python

df_total = df_can[years].sum(axis=0)
df_total = pd.DataFrame(df_total)
df_total.reset_index(inplace=True)
df_total.columns = ['Year','Total']

df_total.plot(kind="scatter", x='Year', y='Total', figsize=(10,8))

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```
